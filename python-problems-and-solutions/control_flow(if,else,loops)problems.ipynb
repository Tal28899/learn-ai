### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Taking input from the user
ctc = int(input("enter your yearly salary : "))

# salary dedection by the company
hra =  (ctc * 10/100)
da =  (ctc * 5/100)
pf  = (ctc * 3/100)

# Convert CTC to lakhs for tax slab comparison
ctc_in_lakhs = ctc / 100000

# Total tax on salary
if ctc_in_lakhs < 5 :
  tax_rate = 0
elif ctc_in_lakhs >= 5 and ctc_in_lakhs < 10:
  tax_rate = 10/100
elif ctc_in_lakhs >= 10 and ctc_in_lakhs < 20:
  tax_rate = 20/100
else: # ctc_in_lakhs >= 20
  tax_rate = 30/100

tax = ctc * tax_rate # Tax is calculated on the full CTC

# total deducted amount
amount_ded = hra + da + pf + tax
# yearly salary
yearly_sal = ctc - amount_ded
# monthly salary
mon_sal =int (yearly_sal / 12)
# printing the results
print(f"your monthly salary is : {mon_sal}pkr")
````

### Code explanation
This code calculates an individual's in-hand monthly salary after various deductions and taxes based on their Cost To Company (CTC).

1.  **Input**: It first takes the yearly CTC as an integer input from the user.
2.  **Company Deductions**: It calculates HRA (10%), DA (5%), and PF (3%) based on the total CTC.
3.  **Tax Slab Conversion**: The `ctc` is converted into 'lakhs' (`ctc_in_lakhs`) to accurately apply the tax slab rates.
4.  **Tax Calculation**: An `if-elif-else` ladder determines the applicable `tax_rate` based on `ctc_in_lakhs`:
    *   Below 5 lakhs: 0% tax
    *   5-10 lakhs: 10% tax
    *   10-20 lakhs: 20% tax
    *   Above 20 lakhs: 30% tax
    The `tax` amount is then calculated by multiplying the full `ctc` by the determined `tax_rate`.
5.  **Total Deductions**: All calculated deductions (`hra`, `da`, `pf`, and `tax`) are summed up to get `amount_ded`.
6.  **Net Salary**: The `yearly_sal` is found by subtracting `amount_ded` from the `ctc`.
7.  **Monthly Salary**: Finally, `mon_sal` is calculated by dividing the `yearly_sal` by 12 and is converted to an integer.
8.  **Output**: The program prints the final calculated `mon_sal`.

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
# making an empty list
fab_series = []
# calculating first 10 num of the series
for i in range(0,10):
  if i < 2:
    fab_series.append(i)

  else:
    fab_num = fab_series[i-2] + fab_series[i-1]
    fab_series.append(fab_num)

print(fab_series)


````

### Code explanation
Here, an empty list `fab_series` is initialized. A `for` loop then iterates 10 times to generate the first 10 numbers of the Fibonacci series:
*   For the first two terms (when `i` is 0 or 1), the numbers (0 and 1) are directly appended to the list.
*   For subsequent terms, the current Fibonacci number (`fab_num`) is calculated by adding the two preceding numbers in the `fab_series` list (`fab_series[i-2]` and `fab_series[i-1]`). This `fab_num` is then appended to the list.
Finally, the complete `fab_series` list containing the first 10 Fibonacci numbers is printed.

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
num = 5
# calulating factorial
fac = 1
for i in range (5,0,-1):
  fac *= i
# printing the result
print(f"factorial of {num} is : {fac}")
````

### Code explanation
Here, a `for` loop is used to calculate the factorial of 5. The loop iterates in reverse order from 5 down to 1 (inclusive), multiplying `fac` by each number. This effectively computes 5 * 4 * 3 * 2 * 1, storing the result in the `fac` variable.

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# taking input from the user
inte = int(input("Enter an integer to reverse: "))
reversed_num = 0
original_num = inte # Store the original number for printing
# revering the number
while inte > 0:
    digit = inte % 10
    reversed_num = reversed_num * 10 + digit
    inte //= 10
print(f"The reverse of {original_num} is: {reversed_num}")
````

### Code explanation for Problem 7
This  code for Problem 7 takes an integer input from the user and reverses its digits.

1.  **Input**: It prompts the user to `Enter an integer to reverse:`.
2.  **Initialization**: It initializes `reversed_num` to `0` and stores the original input in `original_num` for later display.
3.  **Reversal Logic (while loop)**:
    *   The `while` loop continues as long as `inte` (the number being processed) is greater than 0.
    *   Inside the loop, `digit = inte % 10` extracts the last digit of `inte`.
    *   `reversed_num = reversed_num * 10 + digit` builds the reversed number by shifting the existing `reversed_num` one place to the left (multiplying by 10) and adding the extracted `digit`.
    *   `inte //= 10` removes the last digit from `inte` using integer division.
4.  **Output**: Finally, it prints the original number and its reversed version.

### Solution Explanation for Problem 14: Angle Between Hour and Minute Hands

This problem requires calculating the smaller angle between the hour and minute hands of an analog clock at a given time (H hours, M minutes).

Here's the breakdown of the logic:

1.  **Input**: The program takes two integer inputs: `hour` (0-12) and `min` (0-59).

2.  **Normalize Hour**: The `hour = hour % 12` line normalizes the hour. If the hour is 12, it becomes 0 for calculation purposes (representing the 12 o'clock position as 0 degrees).

3.  **Calculate Minute Hand Angle (`minute_angle`)**:
    *   A clock face is 360 degrees.
    *   There are 60 minutes in an hour.
    *   So, the minute hand moves `360 / 60 = 6` degrees per minute.
    *   `minute_angle = min * 6` calculates the angle of the minute hand from the 12 o'clock position.

4.  **Calculate Hour Hand Angle (`hour_angle`)**:
    *   The hour hand moves `360 / 12 = 30` degrees per hour.
    *   It also moves a little bit for every minute that passes. In 60 minutes, the hour hand moves 30 degrees. So, in 1 minute, it moves `30 / 60 = 0.5` degrees.
    *   `hour_angle = (hour * 30) + (min * 0.5)` calculates the angle of the hour hand from the 12 o'clock position. This includes the angle due to the hour itself and the additional movement due to the minutes past the hour.

5.  **Calculate Absolute Difference (`diff`)**:
    *   `diff = abs(hour_angle - minute_angle)` finds the absolute difference between the two calculated angles. This gives one of the two possible angles between the hands.

6.  **Find Minimum Angle (`final_angle`)**:
    *   Because there are always two angles between the hands (e.g., if one is 90 degrees, the other is 270 degrees), we need the *minimum* of these two.
    *   `final_angle = min(diff, 360 - diff)` selects the smaller angle. If `diff` is greater than 180, then `360 - diff` will be the smaller angle.

7.  **Floor the Result**: `math.floor(final_angle)` takes the floor of the final angle, meaning it rounds down to the nearest whole number, as specified in the problem statement.

8.  **Output**: The program then prints this `final_angle`.

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
#taking input from the user
star_value = int(input("enter starting value of the range  :"))
end_value = int(input("enter ending value of the range :"))
for i in range(star_value,end_value + 1):
  num = str(i)
  csum_of_dig = 0
  pow_val = len(num)
  for d in num:
    d = int(d)
    cubed_d = d** pow_val
    csum_of_dig += cubed_d
  if csum_of_dig == i:
      print(f"{i} is a armstrong number.")

````

###`Problem 14`:Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

### Explanation:

1.  **`import math`**: This line imports the `math` module, which contains various mathematical functions, including trigonometric ones.
2.  **`math.pi`**: This is a constant from the `math` module that represents the value of Pi (approximately 3.14159).
3.  **`math.sin(x)`**: This function returns the sine of `x`, where `x` is an angle measured in radians.
4.  **`math.radians(degrees)`**: This function converts an angle from degrees to radians, which is necessary before passing it to `math.sin()` if your input is in degrees.

````python
import math
 # Import builtins to access the original min function

hour = int(input("Enter time in hours (0-12): "))
minute = int(input("Enter time in minutes (0-59): "))

# Ensure hour is within 0-12 range for calculation (12 becomes 0 for 12 o'clock reference)
hour = hour % 12

# Calculate the angle of the minute hand
# 1 minute = 6 degrees (360 degrees / 60 minutes)
minute_angle = minute * 6

# Calculate the angle of the hour hand
# 1 hour = 30 degrees (360 degrees / 12 hours)
# Also, the hour hand moves 0.5 degrees per minute (30 degrees / 60 minutes)
hour_angle = (hour * 30) + (minute * 0.5)

# Calculate the absolute difference between the two angles
diff = abs(hour_angle - minute_angle)

# The angle can be either diff or 360 - diff. We need the minimum of the two.
# Explicitly use builtins.min to avoid shadowing issues
final_angle = min(diff, 360 - diff)

# Print the floor of the final result angle
print(math.floor(final_angle))
````
