### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# importing math libarary to perform calculations
import math
#getting coordinates from user
x1,y1 =map(int,input('enter first vector coordinates (x,y) : ').split(","))
x2,y2 =map(int,input('enter second vector coordinates (x,y) : ').split(","))
# Formula to find eucladian distance
ed = math.sqrt((x2-x1)**2 + (y2-y1)**2)
# printing the results
print(f"eucladian distance is : {ed}")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# taking value from the user
legs = int(input("Enter no of legs : "))
heads = int(input("Enter no of heads : "))

# formula for counting no of chicken and dogs
dogs = (legs - 2*heads)//2
chickens = heads - dogs
# Handling errors
if dogs < 0 or chickens < 0 or legs%2 != 0:
  print('Invalid values for legs and heads')
else:
  #printing the results
  print(f"total no of dogs = {dogs}")
  print(f"total no of chicken = {chickens}")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
#taking nth value from user
Value_n = int(input("Enter a natural number : "))
# sum of all given natural numbers
total_sum = 0 # initial value
# running loop
for x in range(1,Value_n + 1):
   total_sum += x
# calculating mean
mean = total_sum / Value_n

#calculating sum of saqures(ss)
ss = 0
for x in range (1,Value_n + 1):
  saq_diff = (x - mean)**2
  ss += saq_diff
#printing the results
print(f"Sum of saqures error = {ss}")


````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
import math

# Taking numerator and denominator for the first fraction from user
nume1 = int(input("Enter numerator for fraction 1: "))
deno1 = int(input("Enter denominator for fraction 1: "))

# Taking numerator and denominator for the second fraction from user
nume2 = int(input("Enter numerator for fraction 2: "))
deno2 = int(input("Enter denominator for fraction 2: "))

# Calculate a common denominator
common_deno = deno1 * deno2

# Adjust numerators to the common denominator
adjusted_nume1 = nume1 * deno2
adjusted_nume2 = nume2 * deno1

# Sum the adjusted numerators
sum_nume = adjusted_nume1 + adjusted_nume2

# Find the greatest common divisor (GCD) to simplify the fraction
gcd_val = math.gcd(sum_nume, common_deno)

# Simplify the fraction
simplified_nume = sum_nume // gcd_val
simplified_deno = common_deno // gcd_val

# Printing the result
print(f"The sum of the fractions is: {simplified_nume}/{simplified_deno}")
````
