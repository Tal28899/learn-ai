#Strings

String is a datatype in Python. The values inside single quotes ('talha'), double quotes ("talha"), and triple quotes
("""Talha AI Engineer""") are strings in Python.
Triple quotes are used for multi-line strings.

**Memory Representation**

At a fundamental level, strings are sequences of Unicode characters.

```python
print("hello world")
```

#Indexing in Strings

We can access the value in a string through its index. Every value in a string has an index; the first value of a string has zero index, the second has 1 index, and so on.

```python
s ="hello world"
print(s[0])#first value of the string
print(s[1])#second value of the string
```

##Negative indexing

We can also access values of strings through negative indexing. In this method, indexing starts from -1, which corresponds to the last character of the string, and continues backwards.

```python
s = "hello world"
print(s[-1])#last value of the string
print(s[-3])#third last value of the string
```

#slicing in string

It is also used to access parts of a string within a specified range.

```python
s = "Eid Mubarak"
print(s[0:5])#first four element in string
             #space is also given an index value
print(s[0:])#print all elements of a string
print(s[:])#print all elements of a string

print(s[0:8:2])#giving stepsize in range
print(s[8:2:-1])#reversing string through negative step size
```

#Negative Slicing



Here, we access parts of a string using a specified range with negative indices.

```python
s =  "Eid Mubarak"
print(s[-1:-5 :-1])#printing string form last letter
print(s[::-1])#reversing the whole string
```

##3. Logical operators

```python
# and operator
# In Python, non-empty strings are considered 'truthy' and empty strings are considered 'falsy'.
# The 'and' operator returns the first falsy operand it encounters. If all operands are truthy, it returns the last operand.

"Talha" and "engineer" # 'Talha' is truthy, so it evaluates 'engineer'. 'engineer' is truthy, so it returns 'engineer'.
"" and "engineer" # '' is falsy, so it immediately returns ''.
```

```python
# or operator
# The 'or' operator in Python evaluates operands from left to right.
# It returns the first *truthy* operand it encounters. If all operands are *falsy*,
# it returns the last falsy operand.

# Example 1: First operand is truthy
# 'Talha' is a non-empty string, so it's truthy. The 'or' operator returns it immediately.
result1 = "Talha" or "engineer"
print(f"'Talha' or 'engineer' : {result1}")

# Example 2: First operand is falsy, second is truthy
# '' is an empty string, so it's falsy. The 'or' operator then evaluates 'engineer'.
# 'engineer' is a non-empty string, so it's truthy. The 'or' operator returns it.
result2 = "" or "engineer"
print(f"'' or 'engineer' :{result2}")

# Example 3: Both operands are falsy
# '' is falsy. The 'or' operator then evaluates the second ''.
# The second '' is also falsy. Since all operands are falsy, the 'or' operator returns the last one.
result3 = "" or ""
print(f"'' or '' : {result3}")

```

```python
# not operator
not "Talha"#return false if the string is truthy
```

```python
not ""# return true if the string is falsy
```
