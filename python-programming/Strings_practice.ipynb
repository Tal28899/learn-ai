 - Find the length of a given string without using the len() function

```python
profession = "AI Engineer"
length = 0  # Initialize a counter for the length

# Iterate through each letter in the string
for letter in profession:
    length += 1  # Increment the counter for each letter

print(length)
```

 - Extract username from a given email.

```python
mail = "talhafareed0101@gmail.com"
user_name = ""  # Initialize an empty string to store the username

# Iterate through each character in the email string
for letter in mail:
  if letter == '@':  # Stop when the '@' symbol is encountered
    break
  else:
    user_name += letter  # Append the character to user_name

print(user_name)
```

 - Count the frequency of a particular character in a provided string

```python
s = "hello how are you"
char_to_count = 'h'
count = 0  # Initialize a counter for the character frequency

# Iterate through each character in the string
for char in s:
    if char == char_to_count:  # Check if the current character matches the target character
        count += 1  # Increment the counter if there's a match

print(count)
```

 - Write a program which can remove a particular character from a string.

```python
s = 'learn'
upd_s = ''  # Initialize an empty string to build the updated string

# Iterate through each letter in the original string
for letter in s:
  if letter == "l":  # Check if the current letter is the one to be removed
   continue  # Skip this letter if it matches
  else:
    upd_s += letter  # Append the letter to the updated string if it doesn't match

print(upd_s)
```

 - Write a program that can check whether a given string is palindrome or not

```python
# Check if a string is palindrome
s = input("enter your string : ")
rev_s = s[::-1]  # Reverse the string using slicing

# Compare the original string with its reversed version
if s == rev_s:
  print(f"{s} is palindrome")
else:
  print(f"{s} is not palindrome")
```

-  Write a program to count the number of words in a string without split()

```python
s = input("enter your string : " ).strip()  # Get input and remove leading/trailing whitespace

# Handle edge cases: empty string or string with no spaces
if not s:
    total_words = 0
elif ' ' not in s:
    total_words = 1
else:
    total_words = 1  # Start with 1 to count the first word
    # Iterate through the string to count spaces that separate words
    for i in range(len(s) - 1):
        # A word boundary is found when a space is followed by a non-space character
        if s[i] == ' ' and s[i+1] != ' ':
            total_words += 1

print(total_words)
```

 - Write a python program to convert a string to title case without using the title()

```python
s = input("enter your string : " ).lower()  # Get input and convert to lowercase for consistent processing
title_s = ''  # Initialize an empty string for the title-cased result
capitalize_next = True  # Flag to determine if the next letter should be capitalized

# Iterate through each letter in the string
for letter in s:
    # Capitalize the letter if `capitalize_next` is True and it's an alphabetic character
    if capitalize_next and 'a' <= letter <= 'z':
        title_s += letter.upper()
        capitalize_next = False  # Reset flag after capitalizing
    else:
        title_s += letter  # Append the letter as is

    # Set `capitalize_next` to True if the current letter is a space (start of a new word)
    if letter == ' ':
        capitalize_next = True
    # If it's not a space and we just capitalized, ensure next is not capitalized unless a new word starts
    # (This else branch is crucial for consecutive non-space characters after a capitalized one)
    else:
        capitalize_next = False

print(title_s)
```
