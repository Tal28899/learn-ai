## List
A list is a data structure used to store different data types. They are dynamic arrays, meaning they can add items on the fly.

### Example:
l = [1,4,"Hi",[1,2,3],True,5.06]

## Lists vs. Arrays

### Fixed and Dynamic

- Arrays are of fixed size
ex:
`int A[10]` # array capable of holding only 10 integer values
- Lists are dynamic arrays, meaning they can add as many items on the fly.
ex:
`L =[1,3,4]` # list storing three integers

## Homogeneous vs. Heterogeneous

- Arrays are homogeneous, meaning they can only store the same data types.

ex:
`int A[3] = [1,2,3]`

- Lists can be homogeneous (storing the same data types) or heterogeneous (storing different data types).

ex:
`l = [1,2,3]`# homogeneous list
`l = [1,2,3,"hi"]`# heterogeneous list

## Memory

- Lists can occupy more memory than arrays.
- This is because lists are dynamic; when they need to grow beyond their allocated capacity, a new, larger list object is created in memory (often doubling the original size), and the items from the old list are copied to the new one.
- This 'on-the-fly' resizing mechanism and the overhead of storing heterogeneous data types contribute to higher memory usage compared to fixed-size, homogeneous arrays.

## Speed

 - Lists are slower than Arrays.
 - Lists store pointers or references pointing towards original data while Arrays store contiguous blocks of memory.

## Indexing and slicing in list

````python
# indexcing
l = [1,3,4]

# like strings list indexing also starts from zero
print(l[0]) # accessing  first item
print(l[1]) # accessing  second item
print(l[2]) # accessing  third item
````

````python
# negative indexing

l = [1,3,4]

# negative indexing  starts from -1
print(l[-1]) # accessing  last item
print(l[-2]) # accessing  second last item
print(l[-3]) # accessing  third last item
````

````python
# indexcing in a nested list
l=[[[1,2],[3,4]],[[5,6],[7,8]]]
print(l[0][0][1]) # indexing goes from outer list to inner
````

````python
# Slicing
l = ['hi','ai','python',1,'learning']
print(l[0:])# printing all the elements in a list
print(l[0:3]) # printing the firt three elements of list
print(l[::-1]) # inversing the list
print(l[-5:-1]) # negative slicing
````

## Adding items in a list

``` append ```
- It adds elements in the end of the list

````python
l = [1,2,3]
for i in range (4,10):
  l.append(i)
print(l)
````

```extend```
- Breaks iterable into individual elements,add each individual element one by one in the end of the list.


````python
l1 = [1,3,4,5]
l2 = [6,7,5]
l1.extend(l2)
print(l1)
````

``` insert ```
- add elements at a specific index

````python
l1 = [1,3,4,5]
l1.insert(0,90)
print(l1)
````

## Deleting item from a list

 ```del```
 - use  delete the complete list or a part of it.

````python
l = [1,2,3,4,5,6,7,8]
del l[0] # deleting the first element
print(l)
del l[1:4] # deleting a slice of list
print(l)
del l
# print(l) # This would raise a NameError because 'l' no longer exists
````

``` remove ```
- Use to delete a specific item from a list by its value. It removes the first occurrence of that item if there are duplicates.

````python
# Example of remove
fruit_list = ['apple', 'banana', 'orange', 'banana']
fruit_list.remove('banana') # Removes the first occurrence
print(fruit_list)
````

````python
l = [1,2,3,4,5,6,7,8]
l.remove(5)
print(l)
````

``` pop ```
- Use to remove and return the item at a given index. If no index is specified, it removes and returns the last item in the list.

````python
l = [1,2,3,4,5,6,7,8]
l.pop(1) # removes the second element
print(l)
l.pop() # removes the last element
````

``` clear ```
- Clears the list, making it empty.

````python
l = [1,2,3,4,5,6,7,8]
l.clear()
print(l)
````

``` shallow copy ```
- Creates a new object, but for nested objects, it copies references instead of creating new copies of the nested objects.

````python
l = [4,3,5,7,2]
print("ID of original list l:", id(l))
l2 = l.copy() # Correctly perform a shallow copy
print("ID of shallow copy l2:", id(l2))
````

## List comprehension

``` [expression for item  in iterable if condition == True] ```

### Benefits
- Fewer lines of code.
- Time efficient and space efficient.
- Faster than loops.

````python
# printing numbers using loops and storing in list
li=[]
for i in range(1,11):
  li.append(i)
print(li)
````

````python
# printing numbers using  list comprehension
li = [i for i in range(1,11)]
print(li)
````

````python
# printing divible of 5 using list comprehension
l =[i for i in range(1,51) if i % 5 == 0]
print(l)
````

````python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
nes_l=[[i*j for j in range (1,4)] for i in range(1,4)]
print(nes_l)
````

````python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]
cartesian_product = [(i , j) for i in L1 for j in L2]
print(cartesian_product)
````

## zip
- It returns a zip object.
- Pair up elements in order wise of two or more iterables.
- Incase of unequal iterables it pair up elements until  least iterable is completed.

````python
l1= [1,2,3,4]
l2 = [3,4,5,1]
print(list((zip(l1,l2))))
````

````python
# Zip with unequal iterables
names = ["talha", "Ali"]
scores = [85, 90, 95]
# It stops at the shortest list (names)
print(list(zip(names, scores)))
````
