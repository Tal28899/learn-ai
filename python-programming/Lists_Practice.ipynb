###`Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]
```

````python
from itertools import zip_longest

list1 = ["M", "na", "i", "Kh", "ai"]
list2 = ["y", "me", "s", "an"]
n_l= []

# Use zip_longest to combine lists, filling with None for shorter list
# Then filter out None and wrap single elements in lists
for item1,item2 in zip_longest(list1,list2):
  if item1 is not None and item2 is not None:
    n_l.append([item1,item2])
  elif item1 is None:
    n_l.append([item2])
  else:
    n_l.append([item1])


print(n_l)
````

###`Problem 4:` Running Sum on list
Write a program to print a list after performing running sum on it.

i.e:

`Input:`
```
list1 = [1,2,3,4,5,6]
```
`Output:`
```
[1,3,6,10,15,21]
```

````python
list1 = [1,2,3,4,5,6]
run_sum=[]
for i in range(len(list1)):
  if i == 0 :
    run_sum.append(list1[i])
  else :
    run_sum.append(list1[i] + run_sum[i-1])
print(run_sum)
````

````python
from itertools import accumulate

list1 = [1,2,3,4,5,6]

# Use accumulate to get the running sum
run_sum = list(accumulate(list1))
print(run_sum)
````

###`Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
l=[2,4,6,10,1]
n_l = []
for num in l:
  sum = 0
  for nums in l:
    if nums >= num:
      sum += nums
  n_l.append(sum)
print(n_l)
````

###`Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
l = ['1ac21', '23fg', '456', '098d', '1', 'kls']
pro_l = []

for word in l:
    pro = 1
    # Check if there are any digits in the word to handle the product calculation
    for letter in word:
        if letter.isdigit():
          pro *= int(letter)
    pro_l.append(pro)

lis_with_prod = list(zip(l, pro_l))

# Sort the list of tuples based on the product value (descending)
lis_with_prod.sort(key=lambda item: item[1], reverse=True)

# Extracting the sorted strings
sor_l, _ = zip(*lis_with_prod)

print(list(sor_l))
````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
l =['campusxIs', 'bestFor', 'dataScientist']
n_l = []
for word in l:
  words = ''

  for letter in word:
    if letter.isupper():
      words += ' '
    words += letter
  n_l.append(words)

print(n_l)
````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
l1 = [1,2,3,4,5,1]
l2=[2,3,5,7,8]
n_l = []
for n1 in l1:
  if n1 not in n_l:
    n_l.append(n1)
for n2 in l2:
  if n2 not in n_l:
    n_l.append(n2)
print(n_l)
````
